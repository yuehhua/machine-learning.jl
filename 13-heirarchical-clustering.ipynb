# Heirarchical clustering

```julia
using Clustering, Distances, RDatasets, Plots, StatsPlots
```

## Load data

```julia
iris = RDatasets.dataset("datasets", "iris")
first(iris, 6)
```

## Preprocessing

```julia
n = nrow(iris)
D = zeros(n, n)
for i = 1:n
    a = Vector{Float64}(iris[i, 1:4])
    for j = 1:n
        b = Vector{Float64}(iris[j, 1:4])
        D[i, j] = euclidean(a, b)
    end
end
```

## Model

```julia
result = hclust(D, linkage=:single)
```

## Plot

```julia
plot(result)
```

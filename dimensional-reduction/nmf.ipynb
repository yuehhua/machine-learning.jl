# Non-negative matrix factorization

```julia
using NMF, Distances, RDatasets, Plots
```

## Load data

```julia
iris = RDatasets.dataset("datasets", "iris")
first(iris, 6)
```

## Preprocessing

```julia
X = Matrix(iris[:, 1:4])
```

## Model

```julia
k = 3
r = nnmf(X, k)
```

```julia
W = r.W
```

```julia
H = r.H
```

## Plot

```julia
heatmap(W)
```

```julia
heatmap(H)
```

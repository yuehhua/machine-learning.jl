# Heirarchical clustering

```julia
using Clustering, Distances, RDatasets, Plots, StatsPlots
```

## Load data

```julia
iris = RDatasets.dataset("datasets", "iris")
first(iris, 6)
```

## Preprocessing

```julia
D = pairwise(Euclidean(), Matrix(iris[:, 1:4]), dims=1)
```

## Model

```julia
result = hclust(D, linkage=:single)
```

## Plot

```julia
plot(result)
```
